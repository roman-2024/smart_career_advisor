```python
#df = df.drop(columns=["Name"])

# Display unique suggested careers and the associated skills
career_skills = {}

# Loop through each unique career suggestion
for career in data['suuggestedCarrer'].unique():
    # Filter the data for the current career
    career_data = data[data['suuggestedCarrer'] == career]
    
    # Collect the skills associated with the current career
    skills_list = career_data['skills'].dropna().str.split(', ').explode().unique()  # Get unique skills
    
    # Store the skills in the dictionary
    career_skills[career] = skills_list

# Print the skills for each career suggestion
for career, skills in career_skills.items():
    print(f"Skills for {career}:")
    print(", ".join(skills))
    print("\n")
```

```python
# Display unique suggested careers and the associated skills
career_skills = {}

# Loop through each unique career suggestion
for career in data['suuggestedCarrer'].unique():
    # Filter the data for the current career
    career_data = data[data['suuggestedCarrer'] == career]
    
    # Collect the skills associated with the current career
    skills_list = career_data['subjectskilled'].dropna().str.split(', ').explode().unique()  # Get unique skills
    
    # Store the skills in the dictionary
    career_skills[career] = skills_list

# Print the skills for each career suggestion
for career, skills in career_skills.items():
    print(f"subjectskilled for {career}:")
    print(", ".join(skills))
    print("\n")
```

```python
subjectskilled = {
    "Graphic Designer": ["ICT", "Designing"], 
    "Teacher": ["Bangla", "English"], 
    "Accountant": ["Mathematics", "Accounting"], 
    "Data Scientist": ["Mathematics", "Data Analysis"], 
    "Web Developer": ["Web Development", "ICT"], 
    "Mechanical Engineer": ["Engineering", "Physics"], 
    "Digital Marketer": ["Marketing", "Digital Advertising"], 
    "Business Analyst": ["Business Strategy", "Economics"], 
    "Nurse": ["Healthcare", "Biology"], 
    "AI Engineer": ["Artificial Intelligence", "ICT"]
}

# Function to map career to required skills
def map_required_skill(career):
    return ', '.join(subjectskilled.get(career, []))

# Apply the function to update the 'subjectskilled' column based on 'suuggestedCarrer'
data['subjectskilled'] = data['suuggestedCarrer'].apply(map_required_skill)

# Print the updated dataframe (optional, for inspection)
print(data[['suuggestedCarrer', 'subjectskilled']].head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle

# ✅ 1. Data preparation
df = data.copy()
df = df.drop(columns=['Name'])

X = df.drop(columns=['suuggestedCarrer'])
y = df['suuggestedCarrer']

# ✅ 2. Encode categorical features
label_encoder = LabelEncoder()
object_columns = ['Degree', 'skills', 'interested Area', 'Hooby', 'subjectskilled', 'Backgroud']

for col in object_columns:
    X[col] = label_encoder.fit_transform(X[col])

# Encode the target
y = label_encoder.fit_transform(y)

# ✅ 3. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ✅ 4. Train model
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# ✅ 5. Evaluate model
print(f"Training Score: {model.score(X_train, y_train)}")
print(f"Testing Score: {model.score(X_test, y_test)}")

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# ✅ 6. Get Top-2 predicted careers
y_prob = model.predict_proba(X_test)
top_2_preds = np.argsort(y_prob, axis=1)[:, -2:]

for i in range(2):  # Just printing for first 2 samples
    preds = top_2_preds[i]
    probs = y_prob[i, preds]
    
    sorted_order = np.argsort(probs)[::-1]
    career_1 = label_encoder.inverse_transform([preds[sorted_order[0]]])[0]
    career_2 = label_encoder.inverse_transform([preds[sorted_order[1]]])[0]
    prob_1 = probs[sorted_order[0]]
    prob_2 = probs[sorted_order[1]]

    print(f"Sample {i+1}:")
    print(f"1st Predicted Career: {career_1} with Probability: {prob_1:.2f}")
    print(f"2nd Predicted Career: {career_2} with Probability: {prob_2:.2f}")
    print("-" * 40)

# ✅ 7. Save model and label encoder
model_path = 'F:\\PROJECT_PIPELINE_MLOPS\\SMART_CARRER_ADVISOR\\models\\'

with open(model_path + 'logistic_regression_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

with open(model_path + 'label_encoder.pkl', 'wb') as le_file:
    pickle.dump(label_encoder, le_file)

print("✅ Model and Label Encoder saved successfully!")
```
